```python
mcols=['Application_ID','Current City','Python (out of 3)','R Programming (out of 3)','Deep Learning (out of 3)','Degree','Stream','Current Year Of Graduation','Performance_PG','Performance_UG','Performance_12','Performance_10']
wcols=['Application_ID','Current City','PHP (out of 3)', 'MySQL (out of 3)', 'HTML (out of 3)','CSS (out of 3)', 'JavaScript (out of 3)','AJAX (out of 3)', 'Bootstrap (out of 3)', 'MongoDB (out of 3)','Node.js (out of 3)', 'ReactJS (out of 3)', 'Degree','Stream', 'Current Year Of Graduation', 'Performance_PG','Performance_UG', 'Performance_12', 'Performance_10']
```

```python
ml = pd.DataFrame(df1, columns=mcols)
```

```python
web = pd.DataFrame(df1, columns=wcols)
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
le1 = LabelEncoder()
le2 = LabelEncoder()
```

```python
web['Application_ID'] = le1.fit_transform(web.Application_ID)
```

```python
ml['Application_ID'] = le2.fit_transform(ml.Application_ID)
```

```python
web.rename(columns={"Current City":"Current_City","Current Year Of Graduation":"Current_Year_Of_Graduation"},inplace=True)
```

```python
ml.rename(columns={"Current City":"Current_City","Current Year Of Graduation":"Current_Year_Of_Graduation"},inplace=True)
```

```python
web['Current_City'] = le1.fit_transform(web.Current_City)
web['Current_Year_Of_Graduation'] = le1.fit_transform(web.Current_Year_Of_Graduation)

ml['Current_City'] = le2.fit_transform(ml.Current_City)
ml['Current_Year_Of_Graduation'] = le2.fit_transform(ml.Current_Year_Of_Graduation)
```

```python
import numpy as np
```

```python
web['Stream'].fillna("No_Stream", inplace = True)
web['Degree'].fillna("No_Degree", inplace = True)

ml['Stream'].fillna("No_Stream", inplace = True)
ml['Degree'].fillna("No_Degree", inplace = True)
```

```python
web['Performance_10']=web['Performance_10'].str.extract('(\d+)').astype(float)
web['Performance_12']=web['Performance_12'].str.extract('(\d+)').astype(float)
web['Performance_PG']=web['Performance_PG'].str.extract('(\d+)').astype(float)
web['Performance_UG']=web['Performance_UG'].str.extract('(\d+)').astype(float)
```

```python
ml['Performance_10']=ml['Performance_10'].str.extract('(\d+)').astype(float)
ml['Performance_12']=ml['Performance_12'].str.extract('(\d+)').astype(float)
ml['Performance_PG']=ml['Performance_PG'].str.extract('(\d+)').astype(float)
ml['Performance_UG']=ml['Performance_UG'].str.extract('(\d+)').astype(float)
```

```python
web.drop(columns=['Application_ID'],inplace=True)
ml.drop(columns=['Application_ID'],inplace=True)
```

```python
web['Degree'] = le1.fit_transform(web.Degree)
web['Stream'] = le1.fit_transform(web.Stream)

ml['Degree'] = le2.fit_transform(ml.Degree)
ml['Stream'] = le2.fit_transform(ml.Stream)
```
